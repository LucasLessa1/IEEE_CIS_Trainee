# 2º DataSet - Highest Grossers


```python
df2 = pd.read_csv("C:/Users/lucas/Documents/IEEE/HighestGrossers.csv")
df2.head()
```

```python
df2['TOTAL FOR YEAR'] = df2['TOTAL FOR YEAR'].str.replace(',', '')
df2['TOTAL IN 2019 DOLLARS'] = df2['TOTAL IN 2019 DOLLARS'].str.replace(',', '')
df2['TICKETS SOLD'] = df2['TICKETS SOLD'].str.replace(',', '')
df2['TOTAL FOR YEAR'] = df2['TOTAL FOR YEAR'].str.replace('$', '', regex=True)
df2['TOTAL IN 2019 DOLLARS'] = df2['TOTAL IN 2019 DOLLARS'].str.replace('$', '', regex=True)
df2.head()
```

```python
df2['TOTAL FOR YEAR'] = df2['TOTAL FOR YEAR'].astype(float)
df2['TOTAL IN 2019 DOLLARS'] = df2['TOTAL IN 2019 DOLLARS'].astype(float)
df2['TICKETS SOLD'] = df2['TICKETS SOLD'].astype(float)
df2['TOTAL FOR YEAR'] = df2['TOTAL FOR YEAR'].astype(float)
df2['TOTAL IN 2019 DOLLARS'] = df2['TOTAL IN 2019 DOLLARS'].astype(float)


df2=df2.sort_values(by=["TOTAL FOR YEAR","TOTAL IN 2019 DOLLARS","TICKETS SOLD"], ascending=False)
df2
```

```python
distr= df2.groupby("DISTRIBUTOR")["TOTAL FOR YEAR"].sum() 
print(distr)
distr= distr.to_frame()
```

```python
plt.figure(figsize=(40,40))
# plot chart

distr.plot(kind='pie', y = 'TOTAL FOR YEAR', autopct='%1.1f%%', 
 startangle=270, shadow=False, legend = False, fontsize=10)

plt.show()
```

### Podemos perceber nesse gráfico quem mais lucrou ao longo dos anos no universo cinematográfico

```python
gender= df2.groupby("GENRE")["TICKETS SOLD"].sum() 
gender= gender.to_frame()
gender.plot.pie(subplots=True, figsize=(10, 10));
```

### Filmes do tipo "Adventure" tem maior popularidade.

```python
df2.nlargest(30, "TICKETS SOLD")
```
